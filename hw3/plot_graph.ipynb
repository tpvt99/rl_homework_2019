```python
from packaging import version
import tensorflow as tf
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from scipy import stats
import numpy as np
import tensorboard as tb
```

```python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
```

##### CartPole q4

```python
q4_1_1 = EventAccumulator("cs285/data/hw3_q4_1_1_CartPole-v0_24-01-2021_18-38-15")
q4_1_100 = EventAccumulator("cs285/data/hw3_q4_1_100_CartPole-v0_24-01-2021_20-38-31")
q4_10_10 = EventAccumulator("cs285/data/hw3_q4_10_10_CartPole-v0_24-01-2021_21-06-17")
q4_100_1 = EventAccumulator("cs285/data/hw3_q4_100_1_CartPole-v0_24-01-2021_20-59-09")
```

```python
def convert_events_to_df(events):
    events.Reload()
    return pd.concat([pd.DataFrame(events.Scalars(tag), \
                        columns=['wall_time', 'step', tag]).drop(columns=['wall_time']).set_index('step') \
                                 for tag in events.Tags()['scalars']], axis=1)
```

```python
q4_1_1 = convert_events_to_df(q4_1_1)
q4_1_100 = convert_events_to_df(q4_1_100)
q4_10_10 = convert_events_to_df(q4_10_10)
q4_100_1 = convert_events_to_df(q4_100_1)
```

```python
fig, ax = plt.subplots(figsize=(15,10))
q4_1_1.plot(y='Eval_AverageReturn', ax=ax, label='q4_1_1')
q4_1_100.plot(y='Eval_AverageReturn', ax=ax, label='q4_1_100')
q4_10_10.plot(y='Eval_AverageReturn', ax=ax, label='q4_10_10')
q4_100_1.plot(y='Eval_AverageReturn', ax=ax, label='q4_100_1')

ax.set_title('Average Return')
```
